# **Melakukan Pembacaan Data**

```python
import pandas as pd

# membaca databse
data = pd.read_csv("tips.csv")

# mencetak 10 row teratas dari database
display(data.head(10))
```

# **Matplotlib**

## **Scatter Plot**

```python
import matplotlib.pyplot as plt

# scatter plot data day dengan tip
plt.scatter(data['day'], data['tip'])

# menambahkan judul pada plot
plt.title("Scatter Plot")

# setting label untuk axis X dan Y
plt.xlabel('Day')
plt.ylabel('Tip')

# menampilkan hasil plotting
plt.show()
```

## **Tugas**

Tugas : Tampilkan pada sumbu y adalah Tip dan sumbu x adalah smoker, size atau sex !

```python
# scatter plot dengan variable sumbu y adalah tip dan variabel sumbu x adalah
# smoker, size, dan sex

plt.scatter(data['smoker'], data['tip'], label='Smoker', c=data['size'])
plt.scatter(data['sex'], data['tip'], label='Sex', c=data['size'])

# setting label untuk axis X dan Y
plt.xlabel('Sumbu X')
plt.ylabel('Tip')

# menambahkan judul pada plot
plt.title('Tugas Scatter Plot')

# menampilkan legend dan hasil plotting
plt.colorbar()
plt.legend()
plt.show()
```

## **Menambahkan Colorbar**

```python
# Scatter plot antara day dan tip
plt.scatter(data['day'], data['tip'], c=data['size'],
            s=data['total_bill'])

# Menambahkan judul
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.colorbar()

plt.show()

```

## **3 line chart**

```python
# plot garis tip dan size
plt.plot(data['tip'], label='Tip')
plt.plot(data['size'], label='Size')

# Menambahkan judul pada plot
plt.title("3 Line Chart")

# Menamakan sumbu x dan y
plt.xlabel('Number of Day')
plt.ylabel('Tip')

plt.legend()
plt.show()
```

## **Bar Chart**

```python
# Bar chart
plt.bar(data['day'], data['tip'])

# Memberi judul pada plot
plt.title("Bar Chart")

# setting label untuk axis X dan Y
plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
```

## **Histogram**

```python
# histogram dari total_bills
plt.hist(data['total_bill'])

# Menambahkan judul pada plot
plt.title("Histogram")

plt.show()

```

# **Seaborn**

```python
import seaborn as sns

# membuat lineplot
sns.lineplot(x='sex', y='total_bill', data=data)

# setting judul dengan matplotlib
plt.title('Title using Matplotlib Function')

plt.show()
```

## **Tugas**

Lakukan beberapa percobaan pada google colab dan sajkan visualiasinya pada lembar kerja di bawah ini !

```python
# membuat barplot dengan menggunakan libarary seaborn
sns.barplot(x='day',y='tip', data=data,
            hue='sex')

plt.show()
```

```python
# membuat histogram dan kde (kernel density estimation) plot dengan seaborn
sns.histplot(x='total_bill', data=data, kde=True, hue='sex')

plt.show()
```

# **Plotly**

## **Tugas**

### **Scatter Plot**

```python
import plotly.express as px

# scater plot dengan menggunakan plotly
fig = px.scatter(data, x='day', y='tip', color='sex')

fig.show()
```

### **Line Chart**

```python
# membuat line chart dengan plotly
fig = px.line(data, y='tip', color='sex')

# menampilkan hasil
fig.show()
```

### **Bar Chart**

```python
# membuat bar chart dengan plotly
fig = px.bar(data, x='day', y='tip', color='sex')

fig.show()
```

### **Histogram**

```python
# membuat histogram dengan plotly
fig = px.histogram(data, x='total_bill', color='sex')

fig.show()
```

1.   Maplotlib,

2.   Seaborn,
Keunggulan pada library ini adalah tampilan antarmuka yang lebih baik daripada matplotlib, sehingga dapat didapatkan visualisasi yang baik tanpa harus banyak coding

3.   Plotly,
Keunggulan pada library ini adalah didapatkannya hasil visualisasi yang lebih atraktif dan dapat dilakukan kostumisasi, namun variabel variabel seperti nama sumbu x dan sumbu y nya tidak dapat diubah
