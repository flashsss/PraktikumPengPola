## **Estimasi Kepekatan Parametrik**

```python
# contoh parametric probability density estimation
# library yang digunakan

from matplotlib import pyplot
from numpy.random import normal
from numpy import mean
from numpy import std
from scipy.stats import norm

# membangkitkan sample
sample = normal(loc=50, scale=5, size=1000)

# menghitung parameters
sample_mean = mean(sample)
sample_std = std(sample)
print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))

# mendefinisikan distribusi
dist = norm(sample_mean, sample_std)
values = [value for value in range(30, 70)]
probabilities = [dist.pdf(value) for value in values]

# plotting histogram and pdf
pyplot.hist(sample, bins=10, density=True)
pyplot.plot(values, probabilities)
pyplot.show()
```

## **Estimasi Kepekatan Nonparametrik**

```python
#Import Library yang digunakan
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
from numpy import asarray
from numpy import exp
from sklearn.neighbors import KernelDensity

#Contoh kondisi bimodal data
# membangkitkan sample
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample = hstack((sample1, sample2))
# plot histogram
pyplot.hist(sample, bins=50)
pyplot.show()
```

```python
model = KernelDensity(bandwidth=2, kernel='gaussian')
sample = sample.reshape((len(sample), 1))
model.fit(sample)
```

```python
# sample probabilities for a range of outcomes
values = asarray([value for value in range(1, 60)])
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values)
probabilities = exp(probabilities)
```

```python
# plot the histogram and pdf
pyplot.hist(sample, bins=50, density=True)
pyplot.plot(values[:], probabilities)
pyplot.show()
```

## 2.	Pada estimasi kepekatan parametrik, cari distribusi lain selain distribusi normal dan lakukan visualiasi

```python
#Akan digunakan distribusi eksponensial
import matplotlib.pyplot as plt
from numpy.random import exponential
from numpy import mean
from numpy import std
from scipy.stats import expon

# Membangkitkan sampel
sample = exponential(scale=1.0, size=1000)  # Distribusi eksponensial

# Menghitung parameter
sample_mean = mean(sample)
sample_std = std(sample)
print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))

# Mendefinisikan distribusi
dist = expon(scale=sample_mean)  # Distribusi eksponensial dengan rata-rata yang sama
values = [value for value in range(1, 10)]
probabilities = [dist.pdf(value) for value in values]

# Plot histogram dan pdf
plt.hist(sample, bins=20, density=True)
plt.plot(values, probabilities)
plt.title("Estimasi Kepekatan Parametrik dengan Distribusi Eksponensial")
plt.show()
```

```python
from scipy.stats import binom

# Membangkitkan sample dari distribusi binomial
n = 50  # Jumlah percobaan
p = 0.3  # Probabilitas keberhasilan dalam setiap percobaan
sample = binom.rvs(n, p, size=1000)  # Banyak sampel

# Menghitung parameter
sample_mean = sample.mean()
sample_std = sample.std()
print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))

# Mendefinisikan distribusi binomial
dist = binom(n, p)
values = range(10, 40)  # Rentang nilai yang mungkin dalam distribusi binomial
probabilities = [dist.pmf(value) for value in values]

# Plotting histogram
plt.hist(sample, bins=20, density=True)
plt.plot(values, probabilities)
plt.title('Distribusi Binomial')
plt.show()
```

```python
from numpy.random import gamma
from scipy.stats import gamma as gamma_dist

# Membangkitkan sample dari distribusi Gamma
shape = 2.0  # Parameter bentuk (shape)
scale = 1.5  # Parameter skala (scale)
sample = gamma(shape, scale, size=1000)

# Menghitung parameter-parameter
sample_mean = sample.mean()
sample_std = sample.std()
print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))

# Mendefinisikan distribusi Gamma
dist = gamma_dist(shape, scale)
values = [value for value in range(0, 10)]
probabilities = [dist.pdf(value) for value in values]

# Plotting histogram
plt.hist(sample, bins=20, density=True)
plt.plot(values, probabilities)
plt.title('Distribusi Gamma')
plt.show()
```

## 3.	Buat histogram untuk contoh pembangkitan data dengan jumlah peak = 3 (multimodal) kemudian gunakan fungsi kernel density estimation dan lakukan visualisasi

```python
# Library yang diguanakan
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
from numpy import asarray
from numpy import exp
from sklearn.neighbors import KernelDensity

#Ilustrasi kondisi data bimodal dengan tiga ditribusi yang berbeda
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
# membangkitkan sample
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample3 = normal(loc=60, scale=5, size=500)
sample = hstack((sample1, sample2, sample3))
# plot histogram
pyplot.hist(sample, bins=50)
pyplot.show()

model = KernelDensity(bandwidth=2, kernel='gaussian')
sample = sample.reshape((len(sample), 1))
model.fit(sample)

# sample probabilities for a range of outcomes
values = asarray([value for value in range(1, 90)])
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values)
probabilities = exp(probabilities)

# plot the histogram and pdf
pyplot.hist(sample, bins=50, density=True)
pyplot.plot(values[:], probabilities)
pyplot.show()
```
